```python
import ast
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# convert to tags 
```

```python
def convert(obj):
    L=[] 
    for i in ast.literal_eval(obj): # we will get a dictionary each time . 
        L.append(i['name'])
    return L 
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies['cast'][0]
```

```python
# for cast we will be taking the names of the first 3 characters . 
def cast_conv(obj):
    L = []
    counter = 0 
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name']) 
            counter+=1
        else:
            break 
    return L      
```

```python
movies['cast']=movies['cast'].apply(cast_conv)
```

```python
movies.head(2)
```

```python
movies['crew'][0]
```

```python
# for cast we will be taking the names of the first 3 characters . 
def crew_conv(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name']) 
        else:
            continue
    return L 
```

```python
movies['crew']=movies['crew'].apply(crew_conv)
```

```python
movies.head(2)
```

```python
 #converting overview to a list also
movies['overview']=movies['overview'].apply(lambda x:x.split())  
```

```python
movies.head(2)
```

```python
 # now we will be removing the spaces in between cast, crew, genres, keywords 
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text) :
    y =[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```
